You’ll generate random data, analyze it, and visualize the results.

یک دیتافریم با دو ستون "سن" و "درآمد" بسازید و به صورت رندوم 100تا سطر به آن اضافه کنید. سپس اطلاعات زیر را چاپ کنید
1. میانگین سن
2. میانگین درآمد
3. همه اطلاعات آماری مربوط به درآمد
4. نمودار درآمد بر حسب سن
5. هیستوگرام درآمد

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generate random data
np.random.seed(0)
ages = np.random.randint(18, 65, size=100)
salaries = np.random.normal(50000, 15000, size=100)

# 2. Create a DataFrame
df = pd.DataFrame({
    'Age': ages,
    'Salary': salaries
})

# 3. Analyze the data
print("Average age:", df['Age'].mean())
print("Average salary:", df['Salary'].mean())
print("Salary statistics:\n", df['Salary'].describe())

# 4. Visualize the data
plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
plt.show()

plt.hist(df['Salary'], bins=10)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```
