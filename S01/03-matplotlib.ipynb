```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot Example')
plt.show()
```

```python
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.show()
```

```python
plt.bar(x, y)
plt.title('Bar Plot')
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4)
plt.title('Histogram')
plt.show()
```
